```python
import random


class Pokemon:

    def __init__(self, name, type_, hp, attack, defense,
                 speed, experience, moves, level):
        '''initializes the attributes of the Pokemon class'''
        self.name = name
        self.username = None
        self.type_ = type_
        self.hp = hp
        self.attack = attack
        self.defense = defense
        self.speed = speed
        self.experience = experience
        self.moves = moves
        self.level = level

    def update_level(self, exp):
        '''Will update pokemon level which is based on the current
        level plus experience all under a cube root'''
        self.level = (self.level**3+exp)**(1/3)

    def __str__(self):
        '''if the player has entered a username for their pokemon, this
        function returns the pokemons username and level. If they have
        not entered a username, it will return the pokemon name
        and level.'''
        if self.username:
            res = f"This pokemon is named {self.username} and "
            res += f"is at level {self.level}!"
            return res
        res = f"This pokemon is named {self.username} and "
        res += f"is at level {self.level}!"
        return res

    def printing(self, letters):
        letters = list(letters)
        for letter in letters:
            print(letter, end='')
            if DELAY:
                time.sleep(.07)
        print('\n')

    def first_attack(self, comp):
        '''The first attack in a battle is determined by whoever has the
        fastest speed and returns True if the enemy is faster and False
        if you are faster'''
        if CHARACTERS[comp.name]['Speed'] < CHARACTERS[self.name]['Speed']:
            return True
        return False

    def user_choose_move(self, comp):
        '''Prints a string of all of your pokemons moves, then will give
        you an input option to choose your move as long as its your turn.
        If the player entered a valid move, then it will call the deal
        damage function based on the move they chose, otherwise it will
        return a string that says invalid move and will ask for you to
        choose a move again. This function will also return strings that
        state the players username, if they entered one, and the move the
        chose and will also return the health of the enemy after their move.'''
        moves_str = ', '.join(self.moves[:-1])
        moves_str += ' and ' + self.moves[-1]
        self.printing("Your possible moves are " + moves_str)
        while True:
            move = input("Which move do you want to use? ").title()
            if move in self.moves:
                comp.deal_damage(self, move)
                break
#             elif int(move) > 0 and int(move) <= len(self.moves):
#                 print("A", self.moves[int(move)-1])
#                 comp.deal_damage(self, self.moves[int(move)-1])
            else:
                self.printing("Invalid move!")
        if self.username:
            self.printing(f"{self.username} uses {move} against {comp.name}!")
        else:
            self.printing(f"{self.name} uses {move} against {comp.name}!")
        self.printing("Your opponent's health is " + str(comp.hp))

    def comp_choose_move(self, comp):
        '''Has the computer choose a random move from list of possible
        moves for their pokemon. Uses deal_damage function based on move
        selected. Will then return a string that states the computers
        pokemon and the move they chose as well as the players pokemons
        health after the computers move.'''
        move = random.choice(comp.moves)
        self.deal_damage(comp, move)
        if self.username:
            self.printing(f"{comp.name} uses {move} against {self.username}!")
        else:
            self.printing(f"{comp.name} uses {move} against {self.name}!")
        self.printing("Your health is " + str(self.hp))

    def battle(self, comp, game):
        '''automatically prints your opponents starting health. As
        long as both you and your opponents health is greater than 0,
        then it will call the first_attack function. Either the player
        or the computer will have the first attack based on the outcome
        of the first_attack function. If the players pokemon has less
        than or equal zero health, it will return a statement saying
        you have fainted and another string stating your opponents health
        and vice versa when the computer has less than or equal to 0 health.
        If the player beats the computer, it will call the update_level
        function using the computers level of experience. It will also
        heal the computer character because it will be added to your
        list of pokemons.'''
        self.printing(f"Your opponent starts with {comp.hp} health!")
        while self.hp > 0 and comp.hp > 0:
            if self.first_attack(comp):
                self.user_choose_move(comp)
                if comp.hp > 0:
                    self.comp_choose_move(comp)
            else:
                self.comp_choose_move(comp)
                if self.hp > 0:
                    self.user_choose_move(comp)
            print('\n')
        if self.hp <= 0:
            if self.username:
                self.printing(f"{self.username} fainted!")
            else:
                self.printing(f"{self.name} fainted!")
            self.printing(f"{comp.name} still has {comp.hp} health!")
            return False
        if comp.hp <= 0:
            self.printing(f"{comp.name} fainted!")
            if self.username:
                self.printing(f"{self.username} still has {self.hp} health!")
            else:
                self.printing(f"{self.name} still has {self.hp} health!")
            self.update_level(comp.experience)
            self.heal()
            comp.heal()
            return True

    def heal(self):
        '''reupdates self.hp as the Pokemons starting HP value.'''
        self.hp = CHARACTERS[self.name]['HP']

    def calculate_effectiveness_type(self, move):
        '''Generates the factor of the effectivesness of the attack'''
        type_ = 1
        for i in self.moves:
            if i.title() in MOVES_DICTIONARY[move.title()]\
            ['super effective against']:
                type_ *= 2
            elif i.title() in MOVES_DICTIONARY[move.title()]\
            ['not very effective against']:
                type_ /= 2
        return type_

    def get_move_properties(self, move, type_):
        '''returns effectiveness factor and power value of a move'''
        return (type_, MOVES_DICTIONARY[move.title()]['power'])

    def calculate_damage(self, attacker, move_stats):
        '''calculates damage of move using power, effectiveness factor,
        and critical hit factor which is randomly calculated.'''
        print("\n")
        type_, power = move_stats
        rand_temp = random.randint(0, 511)
        if rand_temp < attacker.speed:
            critical = 2
            self.printing("CRITICAL HIT")
        else:
            critical = 1
        rand_val = random.uniform(.85, 1)
        self.hp -= critical * rand_val *\
        type_ * (((2 * attacker.level / 5) + 2) *\
                 power * attacker.attack / self.defense) / 50
        self.hp = round(self.hp)

    def deal_damage(self, other, move):
        '''generates a damage value'''
        self_bonus = self.calculate_effectiveness_type(move)
        move_stats = self.get_move_properties(move, self_bonus)
        self.calculate_damage(other, move_stats)
```

```python
import time
import random


poke_class_dict = {'Pikachu': Pikachu, 'Charizard': Charizard,
                   'Squirtle': Squirtle, 'Jigglypuff': Jigglypuff,
                   'Gengar': Gengar, 'Magnemite': Magnemite,
                   'Bulbasaur': Bulbasaur, 'Charmander': Charmander,
                   'Beedrill': Beedrill, 'Golem': Golem,
                   'Dewgong': Dewgong, 'Hypno': Hypno,
                   'Cleffa': Cleffa, 'Cutiefly': Cutiefly}


class Game:

    def printing(self, letters):
        '''Prints text with a time delay'''
        letters = list(letters)
        for letter in letters:
            print(letter, end='')
            if DELAY:
                time.sleep(.07)
        print('\n')

    def __init__(self):
        '''Initializes Pokemon characters'''
        bulb = Bulbasaur()
        charm = Charmander()
        squir = Squirtle()
        # poke_arr is a list of characters that the user has already collected
        self.poke_arr = [bulb, charm, squir]
        welcome = 'POKÉMON STADIUM'
        self.printing(welcome)
        time.sleep(1)  # creates a one second time delay
        self.printing(f'Welcome to the stadium. You will ' +
                'start with the following Pokémon:')
        self.print_poke_arr()  # assigns a starting character
        self.printing(f'Enter the name of the Pokémon you want ' +
        'to learn more about and press enter when you are ready to ' +
        'start the game.')
        user_interest = input().lower().strip()
        while user_interest:  # tells user about their characters' attributes
            if user_interest == 'bulbasaur' or user_interest == '1':
                self.printing(f'NAME\nBulbasaur\nTYPE\nGrass | Poison' +
                '\n\nBASE STATS\nHP: 45\nAttack: 49\nDefense: 49\nSpeed: 45')
            elif user_interest == 'charmander' or user_interest == '2':
                self.printing(f'NAME\nCharmander\nTYPE\nFire\n\nBASE ' +
                'STATS\nHP: 39\nAttack: 52\nDefense: 43\nSpeed: 65')
            elif user_interest == 'squirtle' or user_interest == '3':
                self.printing(f'NAME\nSquirtle\nTYPE\nWater\n\nBASE ' +
                'STATS\nHP: 44\nAttack: 48\nDefense: 65\nSpeed: 43')
            else:
                self.printing('Please choose one of the following: ' +
                'Bulbasaur | Charmander | Squirtle. Make sure your ' +
                'spelling is correct.')
                user_interest = input().lower().strip()
                if user_interest == 'bulbasaur' or user_interest == '1':
                    self.printing(f'NAME\nBulbasaur\nTYPE\nGrass | ' +
                    'Poison\n\nBASE STATS\nHP: 45\nAttack: 49\nDefense: ' +
                    '49\nSpeed: 45')
                elif user_interest == 'charmander' or user_interest == '2':
                    self.printing(f'NAME\nCharmander\nTYPE\nFire\n\nBASE ' +
                    'STATS\nHP: 39\nAttack: 52\nDefense: 43\nSpeed: 65')
                elif user_interest == 'squirtle' or user_interest == '3':
                    self.printing(f'NAME\nCharmander\nTYPE\nWater\n\nBASE ' +
                    'STATS\nHP: 44\nAttack: 48\nDefense: 65\nSpeed: 43')
            self.printing("Choose the next Pokémon to learn about. " +
            "Press enter when you are ready to move on!")
            self.print_poke_arr()
            user_interest = input().lower().strip()

        self.printing("Now you are ready to name your Pokémon! Enter " +
        "'done' when you are done renaming! Choose your Pokémon to rename:")
        self.print_poke_arr()
        while True:  # renames Pokemon by referring to the array index
            # and changing the original name to the user-inputted name
            user_pokemon = input().lower().strip()
            if user_pokemon == 'bulbasaur' or user_pokemon == '1':
                self.rename_pokemon(self.poke_arr[0])
            elif user_pokemon == 'charmander' or user_pokemon == '2':
                self.rename_pokemon(self.poke_arr[1])
            elif user_pokemon == 'squirtle' or user_pokemon == '3':
                self.rename_pokemon(self.poke_arr[2])
            elif user_pokemon == 'done':
                break
            else:
                self.printing("Invalid pokémon selection. Please try again.")
            self.printing("Enter the next name you wish to change. " +
            "If you are ready to begin the game, enter 'done'!")
            self.print_poke_arr()
        self.printing("Please wait for the game to begin")
        print('.')
        time.sleep(1)
        print('.')
        time.sleep(1)
        print('.')
        time.sleep(1)

    def rename_pokemon(self, poke):
        '''Renames Pokemon and demands a valid input'''
        if poke.username:
            self.printing(f"What do you want the new name of " +
            f"{poke.username} to be?")
        else:
            self.printing(f"What do you want the new name of " +
            f"{poke.name} to be?")
        while True:
            new_name = input()
            try:  # try/except function for if some unusual,
                # unreadable character is input
                poke.username = new_name
                self.printing(f"Your pokémon's new name is {poke.username}!")
                break
            except:
                self.printing("Invalid name. Please try again.")

    def print_poke_arr(self, *poke_arr_inp):
        '''prints Pokemon array for user to see what they have'''
        if poke_arr_inp:
            poke_arr = poke_arr_inp[0]
        else:
            poke_arr = self.poke_arr
        for i in range(len(poke_arr)):
            if poke_arr[i].username:
                self.printing(f"{i+1}  |  {poke_arr[i].username}")
            else:
                self.printing(f"{i+1}  |  {poke_arr[i].name}")

    def select_user_pokemon(self, poke_arr):
        '''Allows the user to choose which Pokemon
        they want to battle with'''
        while True:
            self.printing("Select your pokémon for battle!")
            user_pokemon = input().lower().strip()
            for i in range(len(poke_arr)):
                if poke_arr[i].username:
                    if (user_pokemon == str(i+1) 
                    or user_pokemon == poke_arr[i].username.lower()):
                        poke = poke_arr.pop(i)
                        return poke, poke_arr
                else:
                    if (user_pokemon == str(i+1) 
                    or user_pokemon == poke_arr[i].name.lower()):
                        poke = poke_arr.pop(i)
                        return poke, poke_arr
            self.printing("Invalid Pokémon. Please try again.")

    def generate_opponent(self):
        '''Generates a random Pokemon character as an opponent
        from the Pokemon dictionary'''
        character = random.choice(list(CHARACTERS.keys()))
        comp = poke_class_dict[character.title()]()
        self.printing("Your opponent is " + comp.name)
        return comp

    def battle(self, comp):
        '''Adds a Pokemon if the user wins or takes away a
        Pokemon if the user loses'''
        print("\n")
        poke_arr1 = list(self.poke_arr)
        win = False
        while not win:
            self.print_poke_arr(poke_arr1)
            poke, poke_arr1 = self.select_user_pokemon(poke_arr1)
            print("\n")
            win = poke.battle(comp, self)
            if len(poke_arr1) == 1 and not win:
                self.printing("You suck at this game! Only one " +
                "pokémon left! ")
            elif len(poke_arr1) == 0 and not win:
                self.printing("All your pokémon died!\nGame Over.")
                return False
                break
        if win:
            self.add_pokemon(comp)
            return True

    def add_pokemon(self, comp):
        '''Allows user to decide if they want to add the
        Pokemon to their collection and rename said Pokemon'''
        while True:
            self.printing(f"Do you wish to add {comp.name} " +
            "to your collection? ")
            choice = input()
            if choice.lower() == "yes" or choice.lower() == "y":
                while True:
                    self.printing(f"Do you wish to rename {comp.name}? ")
                    choice = input()
                    if choice.lower() == "yes" or choice.lower() == "y":
                        self.rename_pokemon(comp)
                        break
                    elif choice.lower() == "no" or choice.lower() == "n":
                        break
                    else:
                        self.printing("Unrecognizable input")
                self.poke_arr.append(comp)
                break
            elif choice.lower() == "no" or choice.lower() == "n":
                break
            else:
                self.printing("Unrecognizable input")

    def main(self):
        '''Allows the user to decide if they want to continue 
        and runs the whole code'''
        comp = self.generate_opponent()
        win = self.battle(comp)
        while True:
            if win:
                choice = input("Do you want to play another " +
                "pokemon right now? (Yes/No) ")
                if choice.lower() == "yes" or choice.lower() == "y":
                    comp = self.generate_opponent()
                    self.battle(comp)
                    break
                elif choice.lower() == "no" or choice.lower() == "n":
                    break
                else:
                    self.printing("Unrecognizable input")
            else:
                break


if __name__ == '__main__':
    Game().main()
```
